## Amostragem Estratificada

```python
import pandas as pd
from sklearn.model_selection import train_test_split

iris = pd.read_csv('iris.csv')
# display(iris)
iris['variety'].value_counts()

x, _, y, _ = train_test_split(iris.iloc[:,0:4], iris.iloc[:,4], test_size=0.5, stratify=iris.iloc[:,4])

display(x, _, y, _)
y.value_counts()
```

## Amostragem Sistematica

```python
import numpy as np
import pandas as pd
from math import ceil

populacao = 150
amostra = 15
k = ceil(populacao/amostra)  # ceil serve para arrendondar para cima!!!
print("A amostra é {} ".format(amostra
                               ))
print("A poupulação é {} ".format(populacao))
print("A Proprabilidade é: {} ".format(k))

# Gerando um numero aleatorio inteiro de no minimo 1
r = np.random.randint(low=1, high=k + 1, size=1)

print('O numero sorteado foi: {}'.format(r))

acumulador = r[0]
sorteados = []
for i in range(amostra):
    # print(acumulador)
    sorteados.append(acumulador)
    acumulador += k
    print(acumulador)

base = pd.read_csv('iris.csv')
base_final = base.loc[sorteados]
display(base_final)
```
